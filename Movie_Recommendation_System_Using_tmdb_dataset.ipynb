```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# using apply function 
movies_dataset['genres'] = movies_dataset['genres'].apply(convert)
```

```python
# using apply function on keywords
movies_dataset['keywords'] = movies_dataset['keywords'].apply(convert)
```

```python
# Now again use ast.literal_eval on  'cast' column also in the dataset
def convert_cast(text):
  l = []
  counter = 0
  for i in ast.literal_eval(text):
    if counter < 3:
      l.append(i['name'])
    counter += 1
  return l
 


```

```python
# apply() 
movies_dataset['cast'] = movies_dataset['cast'].apply(convert)
```

```python
movies_dataset.head()
```

```python
movies_dataset['cast'] = movies_dataset['cast'].apply(lambda x:x[0:3])

movies_dataset.head()
```

```python
# fetching th director

def fetch_the_director(text):
  l = []
  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      l.append(i['name'])
  return l
```

```python
movies_dataset['crew'] = movies_dataset['crew'].apply(fetch_the_director)
```

```python
movies_dataset.head()
```

```python
# Replacing all the spaces

def collapse(l):
  l1 = []
  for i in l:
    l1.append(i.replace(" ",""))
  return l1
```

```python
movies_dataset['cast'] = movies_dataset['cast'].apply(collapse)
movies_dataset['crew'] = movies_dataset['crew'].apply(collapse)
movies_dataset['genres'] = movies_dataset['genres'].apply(collapse)
movies_dataset['keywords'] = movies_dataset['keywords'].apply(collapse)
```

```python
# print the first five rows of the dataset

movies_dataset.head()
```

```python
movies_dataset['overview'] = movies_dataset['overview'].apply(lambda x:x.split())
```

```python
movies_dataset['tags'] = movies_dataset['overview'] + movies_dataset['genres'] + movies_dataset['keywords'] + movies_dataset['cast'] + movies_dataset['crew']

movies_dataset.head()
```

```python
new_movies_dataset = movies_dataset.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new_movies_dataset['tags'] = new_movies_dataset['tags'].apply(lambda x: " ".join(x))
new_movies_dataset.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(new_movies_dataset['tags']).toarray()
```

Apply stemming on bag of words using NLTK library

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies_dataset['tags'] = new_movies_dataset['tags'].apply(stem)
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

0 = similarity low

1 = similarity high

```python
similarity  = cosine_similarity(vector)
```

```python
similarity
```

```python
new_movies_dataset[new_movies_dataset['title'] == 'The Lego Movie'].index[0]
```

Recommend top 5 related movies

```python
def recommend(movie):
  movie_index = new_movies_dataset[new_movies_dataset['title'] == movie].index[0]
  distances = sorted(list(enumerate(similarity[movie_index])), reverse= True, key=lambda x:x[1])
  for i in distances[1:6]:
    print(new_movies_dataset.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
